#### Importing libraries and dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('fake_reg.csv')
print(df.shape)
```

```python
df.head()
```

Based on feature 1 and feature 2, we are trying to predict price. This is a supervised learning model

```python
sns.pairplot(df)
```

### Separating feature and target arrays

```python
X = df[['feature1','feature2']].values #we need to pass a numpy array to tensorflow instead of a pandas DataFrame
y = df['price'].values
```

### Splitting arrays in train and validation sets

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.3,random_state=42)
```

### Scaling the feature variables

```python
#We need to scale our features to avoid any issues with the weight and bias terms during model training
from sklearn.preprocessing import MinMaxScaler
#help(MinMaxScaler)
scaler = MinMaxScaler()
scaler.fit(X_train) #to avoid data leakage from val set, we fit the scaler only on training data
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
```

### Building the model

```python
from tensorflow.keras.models import Sequential #to build a base model
from tensorflow.keras.layers import Dense #to add a dense layer on it
```

Method 1 of adding layers: Listing while building the model

```python
#In a sequential model, you can pass a list of multiple dense layers as per your wish to see the network
#Dense has a normal feed forward neural network layer
# model = Sequential([Dense(units=4,activation='relu'),
#                    Dense(units=2,activation='relu'),
#                    Dense(units=1)]) 
```

Method 2 of adding layers: Calling the function of the model to add layers **(preferred)**

```python
#Using this method we can quickly comment out one layer addition line and re-run the model if required

model = Sequential()
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=1)) #last single neuron that gives us a prediction of the price
model.compile(optimizer='rmsprop',loss='mse') #since this is regression problem, we use these parameters for optimization
```

### Choosing an optimizer and loss 
ref:Course resources

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

### Training the model

```python
model.fit(X_train,y_train,epochs=250) #epochs refers to the number of times we will run the model over the entire X_train
#epochs cutoff can be automated later using keras callbacks system based on our target validation loss
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
plt.title("Training Loss per Epoch");
plt.show()
```

As the epochs increase, our loss reduces as show in the above graph, giving us the best fit model

### Evaluating the model

This gives us the loss number i.e. mean squared error in this case on the validation data (unknown data to the model)

```python
model.evaluate(X_val,y_val,verbose=0)
```

Similarly we can check what was the mse on the training set (known data to the model)

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
#Storing validation true and predicted values in a common dataframe

val_pred = model.predict(X_val)
test_predictions = pd.Series(val_pred.reshape(300,))
pred_df = pd.DataFrame(y_val,columns=['Val True Y'])
pred_df = pd.concat([pred_df,test_predictions],axis=1) #if we don't put axis=1 then it concatenates on row
pred_df.columns = ['Val True Y','Val Pred Y']
pred_df
```

```python
# Plotting a scatter plot of true and pred values
sns.scatterplot(x='Val True Y',y='Val Pred Y',data=pred_df)
```

Since this is a dummy dataset, we eventually get a good performing model (seen visually). Let us also measure it using some metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(mean_absolute_error(pred_df['Val True Y'],pred_df['Val Pred Y']))
print(mean_squared_error(pred_df['Val True Y'],pred_df['Val Pred Y']))
```

Our actual dataset has an average price of 498 dollars and this model predicts the prices with an error of 4 dollars

```python
#Importing the model in a new notebook 

later_model = load_model('my_pricemodel.h5')
```

```python
#Using this new loaded model to predict new set of data
later_model.predict(X_val)
```

# Data Pre-Processing

```python
mergeddf = pd.merge(train,first_camp, on=['Patient_ID','Health_Camp_ID'], how='left')
mergeddf2 = pd.merge(mergeddf,second_camp, on=['Patient_ID','Health_Camp_ID'], how='left')
mergeddf3 = pd.merge(mergeddf2,third_camp, on=['Patient_ID','Health_Camp_ID'], how='left')
```

```python
mergeddf3.head()
```

```python
final_train = mergeddf3.copy()
```

```python
final_train.drop(['Donation','Last_Stall_Visited_Number'],axis=1,inplace=True)
```

```python
final_train.head()
```

```python
# def target_func(x,y,z):
#     if (x>0) | (y>0) | (z>0):
#         return 1
#     else:
#         return 0
```

```python
# final_train['Target2'] = target_func(final_train['Health_Score'],final_train['Health Score'],final_train['Number_of_stall_visited'])
```

```python
# final_train['Target2'] = final_train['Health_Score'].apply(lambda x: target_func(x),axis=1)

# final_train['Target2'] = final_train.apply(lambda x: target_func(x['Health_Score'],x['Health Score'],
#                                                                   x['Number_of_stall_visited']),axis=0)
```

```python
final_train.tail()
```

```python
final_train['Outcome'] = np.where((final_train['Health_Score']>0) |
                                 (final_train['Health Score']>0) |
                                 (final_train['Number_of_stall_visited']>0), 
                                 1, 0)
```

```python
final_train.drop(['Health_Score','Health Score','Number_of_stall_visited'],axis=1,inplace=True)
```

```python
final_train.head()
```

```python
print(final_train['Outcome'].value_counts(),'\n\n',final_train['Outcome'].value_counts(normalize=True))
```

```python
final_train.head()
```

```python
final_train.isna().sum()
```

```python
final_train.dropna(axis=0,inplace=True)
```

```python
final_train.isna().sum()
```

**Adding registration month and date column**

```python
#Converting string to date time format
final_train['Registration_Date'] = pd.to_datetime(final_train['Registration_Date'])
```

```python
final_train['Reg_Month'] = pd.DatetimeIndex(final_train['Registration_Date']).month
final_train['Reg_Day'] = pd.DatetimeIndex(final_train['Registration_Date']).day
final_train['Reg_Year'] = pd.DatetimeIndex(final_train['Registration_Date']).year
```

```python
final_train.head()
```

```python
final_train.drop('Registration_Date',axis=1,inplace=True)
```

**Roadmap for pre-processing:** 
Here, we have to merge all the tables in such a way that our train and test tables will respectively have the following columns -
1. Data Merging: Patient ID can be used as the primary key to join all the tables together, except to extract camp details we will have to use the camp ID as the foreign key
    - Patient Profile - All variables from the patient profile dataframe
    - First camp details - yes/no, Donation, camp start date, camp end date, category of the camp, Health_score (target variable)
    - Second camp details - yes/no, camp start date, camp end date, category of the camp, Health_score (target variable)
    - Third camp details - yes/no, last stall visited number, number of stalls visited (target variable)
    
2. Treating null values

3. Treating outliers 

and so on..

```python
final_train.columns
```

```python
X = final_train[['Health_Camp_ID','Var1', 'Var2','Var3', 'Var4', 'Var5','Reg_Month','Reg_Year']]
y = final_train['Outcome']
```

```python
camp_details.head()
```

```python
from sklearn.preprocessing import 
```

```python
final_train.head()
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_val,y_pred)
```
